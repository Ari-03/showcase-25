# add-project-titles

```python
import pandas as pd
import numpy as np
```

```python
titles = pd.read_csv('project-titles.csv', header=0).iloc[1:, :-3]
titles['Domain'] = titles['Group #'].str.split('-').str[0]
titles
```

```python
roster = pd.read_csv('capstone-roster-2023-11-14.csv')
roster
```

```python
roster['Domain'] = roster['Section'].str.split().str[0]
```

```python
titles = titles.merge(roster[['Domain', 'Section']].drop_duplicates(), on='Domain')
titles
```

```python
titles['Mentor'] = titles['Section'].str.extract(r'\((.+)\)')
```

```python
titles['Mentor Cleaned'] = (
    titles['Mentor']
    .str.replace('Rasmus S Nielsen', 'Emily Ramond')
    .str.replace(', ', ' and ')
    .str.replace('Cindy Burke and Ray Major, and Liang Tian', 'Liang Tian and Cindy Burke')
    .apply(lambda s: ' from '.join(s.split(' / ')[::-1]) if '/' in s else s)
)
```

Now, let's add in poster URLs.

```python
posters = pd.read_csv('poster-links.csv')
posters.head()
```

```python
posters['Group #'] = posters['Filename'].str.replace('.pdf', '', regex=False).str.split().str[0]
posters = posters.rename(columns={'URL': 'poster_url'})
```

```python
titles = titles.merge(posters.drop(columns=['Filename']), on='Group #')
```

```python
def fix_emoji(area):
    return area[2:] + ' ' + area[0]
```

```python
def process_row(r):
    names = r["Names"]
    if "," in names:
        last_comma = names.rfind(", ")
        glue = ", and" if names.count(",") >= 2 else " and"
        names = names[:last_comma] + glue + names[last_comma+1:]
    return f'''
    <span class="badge-domain">{r["Group #"]}</span> <b>{r["Project Title (Want to change your title? Email Suraj.)"]}</b>
    <br>{names} (Mentor{"s" if " and " in r["Mentor Cleaned"] else ""}: {r["Mentor Cleaned"]})
    <br><a href="{r['poster_url']}" target="_" class="bg"><b>Poster ü™ß</b></a>
    <br><br>'''

out = '''<br>'''

for i, block in enumerate(titles['Block (see hdsishowcase.com)'].unique()):
    out += f'''
    <br>
    <a name="block-{i + 1}"></a>
    <h3>{block}</h3>
    '''
    for area in titles.loc[titles['Block (see hdsishowcase.com)'] == block, 'Broad Area'].unique():
        out += f'''<h4>{fix_emoji(area)}</h4>'''
        for i, r in titles.loc[titles['Broad Area'] == area].iterrows():
            out += process_row(r)
        out += '<hr>'
        
with open('../index-template.html', 'r') as f:
    so_far = f.read()
    out_final = so_far.replace('{{content-here}}', out)
    with open('../index.html', 'w') as f2:
        f2.write(out_final)
```
